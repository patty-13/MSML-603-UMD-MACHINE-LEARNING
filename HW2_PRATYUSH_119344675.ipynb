### Explanation 
In the following we generate two vectors of random variables with thenegth of $N$ and then vertically stack them (please note that we could have just generated an $2 \times  N$  matrix but we wanted to introduce vstack in numpy that we would use in future) 

```python
import numpy as np
import matplotlib.pyplot as plt 
import random
mean = 0
sigma = 1
N = 100000
a = np.random.normal(mean, sigma, (1,N))
b = np.random.normal(mean, sigma, (1,N))
X = np.vstack((a,b))
```

### Explanation 
We define the following matrix multiplication 
$$
Y = A X 
$$
Where 
$$
A = \begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
    \end{bmatrix}
$$


### Explanation 
In the following we use identity matrix, which results in two indepenednt random variables/processes

```python
A = np.array([[1,0],[0,1]])
Y = A @ X
plt.figure()
plt.scatter(Y[0,:],Y[1,:])
plt.title('scatter plot of two independent random variables')
plt.xlabel('x')
plt.ylabel('y')

x_min = np.min(Y[0,:])
x_max = np.max(Y[0,:])
y_min = np.min(Y[1,:])
y_max = np.max(Y[1,:])

N_hist = 50 
x_bins = np.linspace(x_min, x_max, N_hist)
y_bins = np.linspace(y_min, y_max, N_hist)
plt.figure()
plt.hist2d(Y[0,:],Y[1,:], bins = [x_bins,y_bins], cmap = plt.cm.nipy_spectral)
plt.colorbar()
plt.xlabel('x dimension')
plt.ylabel('y dimension')
plt.title(" 2D Histogram")
plt.figure()
plt.hist(Y[0,:], bins= x_bins)
plt.xlabel('value of x')
plt.ylabel('frequency')
plt.title('marginal histogram for x dimension  ')
```

### Explanation

Let us calculate the variance of each vector component and also calculate the covariance of the the two components. Remember that 
$$
E_X (x ) = \mu_x = \int_{-\infty}^{+\infty} xf_X(x) dx
$$
For our random valiable we find the sample mean which is defined by 
$$
m_X = \frac{1}{N} \sum \limits_{i=1}^N x_i
$$
The variance can be witten as
$$
\sigma_X^2 = var(X) = E_X [(X-\mu_X )^2] = \int_{-\infty}^{+\infty} (x-\mu_X )^2f_X(x) dx
$$
equivalently 
$$
Var_X = \frac{1}{N} \sum \limits_{i=1}^N (x_i-m_X)^2
$$
Equivalently the covarinace is calculated 
$$
cov(X,Y) = E_{XY} [(X-\mu_X )(Y-\mu_y )] = \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty}(x-\mu_X )(y-\mu_Y )f_{XY}(x,y) dxdy
$$
equivalently 
$$
Cov_{XY} = \frac{1}{N} \sum \limits_{i=1}^N (x_i-m_X)(y_i-m_Y)
$$

#Explanation

Now Let us calculate (EXTREMELY INEFFICIENTLY) the mean, variance and covariance of the data we showed above

```python
x = Y[0,:]
y = Y[1,:]
mean_x = 0
mean_y = 0
var_x = 0
var_y = 0
cov_xy = 0
for i in np.arange(N):
  mean_x = mean_x + x[i]
  mean_y = mean_y + y[i]
mean_x = mean_x /N
mean_y = mean_y/N

print('mean of x is equal to', mean_x )
print('mean of y is equal to', mean_y )

for i in np.arange(N):
  var_x = var_x + np.square(x[i]-mean_x)
  var_y = var_y + np.square(y[i]-mean_y)
  cov_xy = cov_xy + (x[i]-mean_x)*(y[i]-mean_y)
var_x = var_x/N
var_y = var_y/N
cov_xy = cov_xy/N

print('Varinace of x is equal to', var_x )
print('Variance of y is equal to', var_y )
print('Covariance of x & y is equal to', cov_xy )
```
